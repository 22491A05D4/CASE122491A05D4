```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
sumanth=pd.read_csv('energy.csv')
print(sumanth)

srh=sumanth[sumanth['State']=='Telengana']
rcb=sumanth[sumanth['State']=='Tripura']

fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'].sum())
diag[1].bar(rcb['Date'],rcb['Consumption'].sum())

CSE4='Kerala'
CSE5='Delhi'
tdp=[CSE4,CSE5]


# Select only the 'Consumption' column and then calculate the sum
srh=sumanth[sumanth['State']=='Kerala']['Consumption'].sum()
rcb=sumanth[sumanth['State']=='Delhi']['Consumption'].sum()

ysr=[srh,rcb]

fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```
